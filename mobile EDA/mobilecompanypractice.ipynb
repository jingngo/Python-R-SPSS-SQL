```python
merged_data = pd.merge(customer_1,customer_2,on='Nickname') #on is optional, default = inner join, right_on, left_on
merged_data.columns
```

```python
merged_data = merged_data.drop(['Password','C.P.'], axis=1)
```

```python
merged_data['Avg'] = (merged_data['HP'] + merged_data['MP']) / 2
```

```python
merged_data.value_counts
merged_data.sort_values(by=['Avg'],ascending = False)
```

```python
merged_data['Avg'].idxmax(axis=0)
```

```python
merged_data_sample = merged_data.sample(frac=0.1)
```

```python
np.corrcoef(merged_data_sample['HP'], merged_data_sample['Defense'])
```

```python
merged_data_sample = merged_data_sample.sort_values(by=['HP'], ascending=False)
merged_data_sample = merged_data_sample.reset_index(drop=True)
```

```python
# ax, fig = plt.subplots(nrows=1,ncols=1)
plt.plot(merged_data_sample['HP'])
plt.show()
```
